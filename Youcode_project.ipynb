```python
'''
Les noms des colonnes sont : Pays, Age, Genre, Dépense des clients.
Les types de donnees sont principalement des objets (strings) :
Cela signifie que les donnees sont soit des chaînes de caracteres,
soit que pandas ne peut pas déterminer le type de données specifique. Seule la colonne Age est de type float.
Non-null : Le total des lignes est de 65. Si le nombre de valeurs non-null est inferieur a 65,
cela indique qu il y a des cellules vides qu il faut traiter.
'''
```

```python
# Conversion de la colonne en type float
data['Dépenses des clients'] = pd.to_numeric(data['Dépenses des clients'], errors='coerce')
# Remplacement des valeurs NaN par 0
data['Dépenses des clients'] = data['Dépenses des clients'].fillna(0)
```

```python
data
```

```python
mean_age=data['Age'].mean()
med_age=data['Age'].median()
mean_dep=data['Dépenses des clients'].mean()
med_dep= data['Dépenses des clients'].median()
```

```python
print('La moyenne age est {} et la median age est {}'.format(mean_age  , med_age))
print('La moyenne Dépenses des clients est {} et la median Dépenses des clients est {}'.format(mean_dep  , med_dep))
```

```python
### visualisation des données  qui montre les dépenses des clients par pays
plt.figure(figsize=(10,6))
plt.bar(data['Pays'] , data['Dépenses des clients'])
plt.title('les dépenses des clients par pays')
plt.xlabel('pays')
plt.ylabel('Depense des clients')
```

```python
data = data[(data['Dépenses des clients']>10)]  ### eliminer les depense de client moins de 10
data_dup = data.duplicated(keep=False)          ### eliminer les duplicates
data_final= data[~data_dup]
```
